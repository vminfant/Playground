## Finding the probability of getting a head when flipping a coin using Relative Frequency Approach

###### Import libraries

```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

###### Constants 

```python
TRAILS = 10000
```

###### Flip a coin 
* Choose a random number from (0 or 1). 0 indicates a tail and 1 indicates a head

```python
def flip_coin():
    return random.randint(0,1)
```

###### Run trails & calculate the overall probability

```python
def run_trails(trails=TRAILS):
    
    head_count = 0
    tail_count = 0
    results = {}
    for trail in range(1,trails+1):
        result = flip_coin()
        if result: # 1 - indicates a head
            head_count += 1
        else:
            tail_count += 1
            
        results[trail] = [(head_count/trail)*100,(tail_count/trail)*100]
        
    print (f"Probability of getting a head is {(head_count/trails)*100:2.2f}")
    return results
```

##### Run 10000 trails

```python
results = run_trails()
```

```python
trails_df = pd.DataFrame.from_dict(results,orient='index').reset_index()
trails_df.columns = ['trail','head','tail']
```

```python
plt.figure(figsize=(16, 6))
ax_ps = sns.lineplot(x="trail", y="head", color="green",data=trails_df)
```

```python
plt.figure(figsize=(16, 6))
ax_pns = sns.lineplot(x="trail", y="tail", color="orange",data=trails_df)
```
